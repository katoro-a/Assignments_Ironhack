```python
#check the link above to look at more examples of how to apply snake case   
def change_case(str):
    res = [str[0].lower()]
    for c in str[1:]:
        if c in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
            res.append('_')
            res.append(c.lower())
        else:
            res.append(c)
     
    return ''.join(res)
```

```python
customer_df.columns=list(map(change_case,customer_df.columns)) 


#data['timelag_log']= list(map(log_transfom_clean_, data['TIMELAG']))
#data['timelag_log'].hist()
```
