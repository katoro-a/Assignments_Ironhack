# More efficinet way (json_normalize)
- use json_normalize to flatten the file more efficiently 
- unlike the above where you have to do it one by one  
- this is how to manipulate the json data you get from your api
- you will be working with lists and dictionaries 

```python
from pandas.io.json import json_normalize
response = requests.get('https://api.github.com/events')
results = response.json()
flattened_data = json_normalize(results) # this will automatically flatten it (json_normalize)
```

```python
flattened_data = pd.DataFrame(flattened_data)
flattened_data.head()
```

# ACCESSING YOUR OWN PAGE MAKING A SESSION 

```python
import requests
import json
import getpass

username = 'katoro-a'
token = getpass.getpass(prompt='Token: ', stream=None)

# Create a re-usable session object with the user creds in-built
gh_session = requests.Session()
gh_session.auth = (username, token)
```

```python
# get the repos for this particular account 
json.loads(gh_session.get('https://api.github.com/users/katoro-a/repos').text)
# put in your url
```
